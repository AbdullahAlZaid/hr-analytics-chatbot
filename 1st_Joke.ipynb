```python
import requests
```

```python
def get_joke():
    url = "https://official-joke-api.appspot.com/random_joke"

    # 1) Send the request (GET)
    response = requests.get(url, timeout=10)

    # 2) Check if it worked (status code 200 = OK)
    if response.status_code != 200:
        return "Sorry, I couldn't reach the joke service right now."

    # 3) Convert JSON text → Python dictionary
    data = response.json()

    # 4) Extract fields from the JSON
    setup = data.get("setup", "No setup found.")
    punchline = data.get("punchline", "No punchline found.")

    # 5) Return a formatted joke string
    return f"{setup}\n{punchline}"
```

```python
print("JokeBot: Type 'joke' to get a joke, or 'bye' to exit.")

while True:
    user = input("You: ").strip().lower()

    if user == "bye":
        print("JokeBot: Goodbye!")
        break

    elif user == "joke":
        print("JokeBot:", get_joke())

    else:
        print("JokeBot: I only understand 'joke' or 'bye'.")
```
