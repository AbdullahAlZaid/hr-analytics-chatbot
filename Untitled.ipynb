```python
import sqlite3
import pandas as pd

# Create/connect to a local SQLite database file
conn = sqlite3.connect("hr.db")

# Insert the DataFrame into a SQL table called 'employees'
df.to_sql("employees", conn, if_exists="replace", index=False)

# Quick check: count rows inside the SQL table
pd.read_sql_query("SELECT COUNT(*) AS rows_in_db FROM employees;", conn)
```

```python
pd.read_sql_query("""
SELECT Department, COUNT(*) AS Employees
FROM employees
GROUP BY Department
ORDER BY Employees DESC;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
  Attrition,
  COUNT(*) AS Employees,
  ROUND(100.0 * COUNT(*) / (SELECT COUNT(*) FROM employees), 2) AS Percentage
FROM employees
GROUP BY Attrition;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
  Department,
  ROUND(AVG(MonthlyIncome), 2) AS AvgMonthlyIncome
FROM employees
GROUP BY Department
ORDER BY AvgMonthlyIncome DESC;
""", conn)
```
